```python
# Plotting graph of cost per month of the standard plan against the countries

fig = px.pie(df, values='Cost Per Month - Basic ($)', names='Country')
fig.update_traces(textposition='inside')
fig.update_layout(uniformtext_minsize=12, uniformtext_mode='hide')
fig.show()
```

```python
# Plotting graph of cost per month of the basic plan against the countries

fig = px.scatter(df, x='Country', y='Cost Per Month - Premium ($)')
fig.show()
```

```python
# Plotting graph of library size against the countries

fig = px.bar(df, x='Country', y='Total Library Size', 
             hover_data=['No. of TV Shows', 'No. of Movies'], 
             color='Total Library Size', height=600)
fig.show()
```

```python
# Calculating the average subscription charges Netflix earns from respective countries
totalCharges = df['Cost Per Month - Basic ($)']+df['Cost Per Month - Premium ($)']+df['Cost Per Month - Standard ($)']
df['Avg Subscription Charges'] = totalCharges/3
```

```python
df.head()
```

```python
# Sorting the countries from highest subsciption to lowest subcription

df_train = df.sort_values(by='Avg Subscription Charges', ascending=True)
```

```python
df_train.head()
```

```python
df_train.tail()
```

```python
%matplotlib inline
plt.rcParams['figure.figsize']=(12, 6)
Top10 = df_train.groupby('Country')['Avg Subscription Charges'].sum().sort_values(ascending=True)
Top10[:10].plot(kind='bar')
```

```python
%matplotlib inline
plt.rcParams['figure.figsize']=(12, 6)
Top10 = df_train.groupby('Country')['Avg Subscription Charges'].sum().sort_values(ascending=False)
Top10[:10].plot(kind='bar')
```

```python
# Relationship between the countries and the total library size

plt.figure(figsize=(16, 12))
sns.scatterplot(x=df_train['Cost Per Month - Basic ($)'], y=df_train['Country'], size=df_train['Total Library Size'])
plt.xlabel('Cost Per Month - Basic ($)')
plt.ylabel('Country')
```

```python
# Relationship between the cost per month of standard plan and the total library size
plt.figure(figsize=(16, 12))
sns.scatterplot(x=df_train['Total Library Size'], y=df_train['Cost Per Month - Standard ($)'], size=df_train['Total Library Size'])
plt.xlabel('Total Library Size')
plt.ylabel('Cost Per Month - Standard ($)')
```

```python
# Distribution of the TV shows

plt.figure(figsize=(12,6))
plt.title('Distribution of TV Shows')
sns.distplot(df['No. of TV Shows'], color='green')
plt.show()
```

```python
# Distribution of the movies

plt.figure(figsize=(12,6))
plt.title('Distribution of Movies')
sns.distplot(df['No. of Movies'], color='red')
plt.show()
```

```python
# Multiple visualisations on the same graph

f, ax = plt.subplots(figsize=(6, 18))

dfSorted = df.sort_values('Cost Per Month - Premium ($)', ascending=True)

# Plot the subsciption charges of countries in descending order
sns.set_color_codes('pastel')
sns.barplot(x='Cost Per Month - Premium ($)', y='Country', data=dfSorted,
            label='Premium', color='r')

sns.barplot(x='Cost Per Month - Standard ($)', y='Country', data=dfSorted,
            label='Standard', color='b')

sns.barplot(x='Cost Per Month - Basic ($)', y='Country', data=dfSorted,
            label='Basic', color='g')

# Add a legend and informative axes
ax.legend(bbox_to_anchor=(1.05, 0.3), loc='upper left', borderaxespad=0)
ax.set(title='Netflix Subscription Charges Per Month In Different Countries',
       xlabel='Dollars ($)', xlim=(0, 30))
sns.despine(left=True, bottom=True)
```
